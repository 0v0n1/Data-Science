```python
!pip install spacy pandas
!python -m spacy download en_core_web_md

import spacy
import pandas as pd

# Load spaCy medium model
nlp = spacy.load("en_core_web_md")

# Store results of entities
results = []

print("Named Entity Recognition Tool")
print("Type a sentence and press Enter.")
print("Type 'exit' to finish and save results to CSV.\n")

while True:
    sentence = input("Enter sentence: ").strip()

    if sentence.lower() == 'exit':
        break

    doc = nlp(sentence)
    for ent in doc.ents:
        results.append({
            "Sentence": sentence,
            "Entity": ent.text,
            "Label": ent.label_
        })

    if doc.ents:
        print("\nEntities found:")
        for ent in doc.ents:
            print(f" - {ent.text} ({ent.label_})")
    else:
        print("No named entities found.")
    print()

# Save to CSV
if results:
    df = pd.DataFrame(results)
    csv_name = "ner_results.csv"
    df.to_csv(csv_name, index=False)
    print(f"\nResults saved to '{csv_name}'")
else:
    print("\nNo entities were processed. CSV not saved.")
```
